```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df_2015 = pd.read_csv("2015_.csv")
df_2016 = pd.read_csv("2016_.csv")
df_2017 = pd.read_csv("2017_.csv")
df_2018 = pd.read_csv("2018_.csv")
df_2019 = pd.read_csv("2019_.csv")

# Add year columns
dfs = [df_2015, df_2016, df_2017, df_2018, df_2019]
years = range(2015, 2020)
for df, year in zip(dfs, years):
    df['Year'] = year

# Combine datasets
df = pd.concat(dfs, ignore_index=True)

# 3. Initial Data Exploration
print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
display(df.head())
print("\nLast few rows:")
display(df.tail())

# 4. Summary of Data
print("\nDataset Information:")
df.info()
print("\nSummary Statistics:")
display(df.describe())

# 5. Check for Missing Values
print("\nMissing Values Count:")
display(df.isnull().sum())
print("\nMissing Values Percentage:")
display((df.isnull().sum() / len(df) * 100).round(2))

# 6. Handle Missing Data
# Fill numeric columns with median
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# 7. Check for Duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()

# 8. Rename Columns (if needed)
# Example (adjust according to your column names):
# df = df.rename(columns={'old_name': 'new_name'})

# 9. Identify Outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numeric_columns])
plt.xticks(rotation=45)
plt.title('Box Plot of Numeric Variables')
plt.show()

# 10. Data Transformation
# Example for skewed columns (adjust based on your data):
for col in numeric_columns:
    if df[col].skew() > 1:
        df[f'{col}_log'] = np.log1p(df[col])

# 11. Data Visualization
plt.figure(figsize=(15,12))
sns.histplot(data=df, x='Happiness Score', bins=30)
plt.title('Distribution of Happiness Score')
plt.show()

# 12. Correlation Analysis
plt.figure(figsize=(15, 12))
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# 13. Pair Plot
# Select a subset of columns to avoid overcrowding
important_vars = ['Happiness Score', 'GDP per capita', 'Social support', 'Year']
sns.pairplot(df[important_vars])
plt.show()

# 14. Grouping and Aggregation
region_happiness = df.groupby('Region')['Happiness Score'].agg(['mean', 'count', 'std']).round(3)
print("\nHappiness Score by Region:")
display(region_happiness.sort_values('mean', ascending=False))

# 15. Draw Insights
yearly_happiness = df.groupby('Year')['Happiness Score'].mean()
plt.figure(figsize=(15 10))
yearly_happiness.plot(kind='line', marker='o')
plt.title('Average Happiness Score Over Years')
plt.grid(True)
plt.show()
```
